```python
h_steps, w_steps = w_sol.shape

res = np.zeros(img.shape)

for i in range(0, img.shape[0], h_steps):
    for j in range(0, img.shape[0], w_steps):
        img_crop = img[i:i+h_steps, j:j+w_steps]
        
        try:
            res[i:i+h_steps, j:j+w_steps] = img_crop - w_sol
        except ValueError:
            pass

res[res < 0] = 0
```
