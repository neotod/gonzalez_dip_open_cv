```python
def genereate_zone_plate(size, distance_to_zero=0.8):
    img = np.zeros((size, size))
    
    z = lambda x, y: 5 * (1 + np.cos(x**2 + y**2))
    mid = size // 2
    
    for i in range(size):
        for j in range(size):
            x, y = mid - i, mid - j
            
            dist_to_cent = abs(mid - abs(x)) + abs(mid - abs(y))
            if dist_to_cent > distance_to_zero * mid:
                img[i, j] = z(x, y)
                
            else:
                img[i, j] = 0
                
            
    return img
    
```
