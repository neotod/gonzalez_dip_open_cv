```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from scipy.ndimage import convolve
```

```python
skel_img = cv2.cvtColor( cv2.imread('./skel.png'), cv2.COLOR_BGR2GRAY)
```

```python
plt.gray()
```

```python
plt.imshow(skel_img, cmap='gray')
```

```python
h_filtr = [
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1],
]

v_filtr = [
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1],
]

h_sobel = convolve(skel_img, h_filtr)
v_sobel = convolve(skel_img, v_filtr)
```

```python
skel_img.shape
```

```python
c = 0.02

plt.figure(figsize=(skel_img.shape[0]*c, skel_img.shape[1]*c))
plt.imshow(h_sobel)
```

```python
plt.figure(figsize=(skel_img.shape[0]*c, skel_img.shape[1]*c))
plt.imshow(v_sobel)
```

```python
from scipy.ndimage import gaussian_filter
```

```python
g = np.sqrt(v_sobel**2 + h_sobel**2)
```

```python
plt.figure(figsize=(skel_img.shape[0]*c, skel_img.shape[1]*c))
plt.imshow(g)
```

```python
lena = cv2.cvtColor( cv2.imread('../lena.png'), cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(lena, cmap='gray')
```

```python
lena_g_x = convolve(lena, h_filtr)
lena_g_y = convolve(lena, v_filtr)
```

```python
plt.hist(lena_g_x.ravel());
```

```python
plt.imshow(lena_g_x, cmap='gray')
```

```python
plt.imshow(lena_g_y, cmap='gray')
```

```python
lena_g = np.sqrt(lena_g_x**2 + lena_g_y**2)
```

```python
plt.imshow(lena_g, cmap='gray')
```

```python
lena_g *= 255.0 / lena_g.max()
```

```python
plt.imshow(lena_g, cmap='gray')
```
