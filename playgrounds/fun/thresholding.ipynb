```python
import cv2
```

```python
c1 = cv2.cvtColor( cv2.imread('./coins_1.png'), cv2.COLOR_BGR2GRAY)
c2 = cv2.cvtColor( cv2.imread('./coins_2.png'), cv2.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt
plt.gray()
```

```python
plt.imshow(c1)
```

```python
plt.imshow(c2)
```

```python
(T, c1_th) = cv2.threshold(c1, 248, 255, cv2.THRESH_BINARY_INV)
plt.imshow(c1_th)
```

```python
(T, c2_th) = cv2.threshold(c2, 248, 255, cv2.THRESH_BINARY_INV)
plt.imshow(c2_th)
```

```python
c1.shape
```

```python
plt.imshow(cv2.bitwise_and(c1, c1, mask=c1_th))
```

```python
plt.imshow(cv2.bitwise_and(c2, c2, mask=c2_th))
```
