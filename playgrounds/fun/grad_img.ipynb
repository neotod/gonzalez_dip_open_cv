```python
def gen_grad_img_2d(size, direc):
    if direc == 'horiz':
        row = np.linspace(0, 255, size)
        res = np.tile(row, (size, 1))
    elif direc == 'vertic':
        col = np.linspace(0, 255, size).reshape(size, 1)
        res = np.tile(col, (1, size))

    plt.gray()
    
    return res

def gen_grad_img_3d(size):
    row = np.linspace(0, 255, size)

    plt.gray()
    res = np.tile(row, (size, 1))
    
    return res
```

```python
fig, ax = plt.subplots(1, 5, figsize=(5*5, 5))

ax[0].imshow(gen_grad_img_2d(5, 'horiz'))
ax[1].imshow(gen_grad_img_2d(10, 'horiz'))
ax[2].imshow(gen_grad_img_2d(25, 'horiz'))
ax[3].imshow(gen_grad_img_2d(50, 'horiz'))
ax[4].imshow(gen_grad_img_2d(100, 'horiz'))
```

```python
fig, ax = plt.subplots(1, 5, figsize=(5*5, 5))

ax[0].imshow(gen_grad_img_2d(5, 'vertic'))
ax[1].imshow(gen_grad_img_2d(10, 'vertic'))
ax[2].imshow(gen_grad_img_2d(25, 'vertic'))
ax[3].imshow(gen_grad_img_2d(50, 'vertic'))
ax[4].imshow(gen_grad_img_2d(100, 'vertic'))
```

```python
def get_gradient_3d(width, height, start_list, stop_list, is_horizontal_list):
    result = np.zeros((height, width, len(start_list)), dtype=np.float)

    for i, (start, stop, is_horizontal) in enumerate(zip(start_list, stop_list, is_horizontal_list)):
        result[:, :, i] = get_gradient_2d(start, stop, width, height, is_horizontal)

    return result
```
